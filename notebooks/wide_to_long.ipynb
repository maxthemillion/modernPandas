# Wide to long format (and vice versa) - .stack() .unstack()

```python
import pandas as pd
import seaborn as sns
```

We read a dataset that is in long format. To move from the long format to wide, we first create a MultiIndex on the data frame and then use .unstack() to switch one level of the MultiIndex to columns.

```python
df = sns.load_dataset("flights").set_index(["year", "month"])
```

```python
df.head()
```

```python
df.unstack()
```
